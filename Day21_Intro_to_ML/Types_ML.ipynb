### <u>THE THREE BRANCJES OF MACHINE LEARNING</u>:



### <u>SUPERVISED LEARNING</u>:
Supervised Learning is a type of Machine Learning where the data already has correct answers, called labels. The model learns by studying input and output pairs and tries to understand the relationship between them. After training, it can predict the correct output for new data.

Example: Spam detection is a supervised learning problem because emails are already labeled as “spam” or “not spam.” Another example is predicting house prices using past data where the prices are already known.

### <u>UNSUPERVISED LEARNING</u>:
 Unsupervised Learning is used when the data does not have labels. The model tries to find patterns, similarities, or groups in the data by itself. It explores the data and discovers hidden structures without knowing the correct answers in advance.

Example: Customer segmentation is an example of unsupervised learning, where customers are grouped based on their buying behavior. Another example is grouping similar news articles together based on content.

### <u>Reinforcement Learning</u>:
Reinforcement Learning is a type of Machine Learning where the model learns through trial and error. It receives rewards for correct actions and penalties for wrong ones. Over time, it improves its performance by trying to maximize rewards.

Example: A game-playing AI learns better strategies by playing many times and receiving rewards when it wins. Another example is a robot learning how to walk by adjusting its movements based on feedback.

### <u>Why Choosing the Wrong Type is a Mistake</u>
If we choose the wrong type of learning, the model will not give accurate or meaningful results. Each type of Machine Learning is designed to solve a specific kind of problem, so using the wrong one can create confusion and errors. 
For example:
supervised learning requires labeled data, and without labels, the model cannot learn correctly. On the other hand, unsupervised learning is mainly used to find patterns or group data, so it cannot be used when we need exact predictions. Reinforcement learning is useful only in situations where actions are rewarded or punished, such as in games or robotics. If we apply reinforcement learning to a normal prediction problem, it will not work properly. Choosing the wrong method can waste time, effort, and resources. It may also lead to poor performance and low accuracy. Therefore, it is very important to clearly understand the problem and the type of data available before selecting a Machine Learning approach. Making the right choice helps in building an effective and successful model.
